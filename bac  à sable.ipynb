```python
r_object_center = []
r_object_edge = []
center_mean = []
center_std = []
edge_mean = []
edge_std = []
for iii in np.arange(5):
    print(iii)
    r_object_center.append(rp.raman_mapping_xy(r"P{:d}_center_01.txt".format(iii+1)))
    r_object_center[iii].strain_biax()
    r_object_center[iii].plot_biax()
    center_mean.append(r_object_center[iii].eps_biax_mean)
    center_std.append(r_object_center[iii].eps_biax_std)
    r_object_edge.append(rp.raman_mapping_xy(r"P{:d}_edge_01.txt".format(iii+1)))
    r_object_edge[iii].strain_biax()
    r_object_edge[iii].plot_biax()
    edge_mean.append(r_object_edge[iii].eps_biax_mean)
    edge_std.append(r_object_edge[iii].eps_biax_std)
```

```python
plt.figure()
plt.errorbar(np.arange(5), center_mean, center_std,label = 'center',marker='o')
plt.errorbar(np.arange(5), edge_mean,edge_std, label='edge',marker = 'o')
plt.legend()
plt.show()
```
